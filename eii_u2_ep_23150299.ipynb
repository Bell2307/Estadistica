**Hipótesis**

"El riesgo de hipertensión arterial está influenciado por el nivel de glucosa en sangre, el índice de masa corporal y la edad."

Variables:
Variable dependiente: riesgo_hipertension


Variables independientes:
glucose (nivel de glucosa en sangre),
masa_corporal (índice de masa corporal),
edad

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['masa_corporal', 'medida_cintura', 'peso', 'edad']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```

El coeficiente R² nos ayuda a evaluar qué tan bien el modelo de regresión captura la relación entre las variables independientes y la dependiente.
Un R² alto es deseable, pero no es el único criterio de un buen modelo. También hay que revisar los supuestos estadísticos (como harás en el inciso f).

```python
b0, b1, b2, b3, b4= modelo.params

Fun = lambda X1, X2, X3, X4 : b0 + b1 * X1 + b2 * X2 + b3 * X3 +b4 +X4

Fun(0, 2.5, 4, 1)
```

```python
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(modelo.resid, X_train_constante)
print(f'Valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}')
```

```python
from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'peso ~ edad + masa_corporal +  medida_cintura + concentracion_hemoglobina', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)
tabla_anova
```

**Conclusión final**

El modelo construido confirma que la edad, el nivel de glucosa en sangre y el índice de masa corporal son factores relevantes en la predicción del riesgo de hipertensión arterial. Aunque el modelo explica una parte significativa de la variabilidad, se recomienda considerar en futuros estudios otras variables como antecedentes familiares, nivel de actividad física y factores psicológicos para mejorar la capacidad predictiva.

Este análisis respalda la importancia de monitorear y controlar la glucosa, mantener un peso saludable y prestar atención al envejecimiento como estrategias clave en la prevención de la hipertensión.
